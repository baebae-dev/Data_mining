```python
new = hds_train[['주구매시간','저녁구매비율']]

nonan = new[pd.notnull(new['주구매시간']) == True]
nan = new[pd.notnull(new['주구매시간']) != True]
print(nonan.head())
print(nan.head())
```

```python
nonan_new = nonan.apply(lambda x: round(x,2)) ; nonan_new.head()
```

```python
stand = nonan_new.groupby(['저녁구매비율'])['주구매시간'].mean().reset_index() ; stand.head()
```

```python
stand_apply = hds_train[pd.notnull(hds_train['주구매시간']) != True]['저녁구매비율'].reset_index().apply(lambda x: round(x,2))
stand_apply.head()
```

```python
a = pd.merge(stand_apply,stand,on='저녁구매비율') ;a.head()
```

```python
import math
a['주구매시간'] = a['주구매시간'].apply(lambda x: math.ceil(x))
a.head()
```

```python
for i in range(0,len(a)):
    num = hds_train['custid'][a['index'][i]]
    hds_train.loc[num,'주구매시간'] = a['주구매시간'][i]
```

```python
hds_train=hds_train.fillna({'주구매시간': 18})
hds_train
```

```python
hds_test[pd.notnull(hds_test['주구매시간']) != True].info() # test데이터에 주구매시간 2488개의 결측값 
```

```python
test_stand_apply = hds_test[pd.notnull(hds_test['주구매시간']) != True]['저녁구매비율'].reset_index().apply(lambda x: round(x,2))
test_stand_apply.head()
```

```python
a_test = pd.merge(test_stand_apply,stand,on='저녁구매비율').apply(lambda x: round(x,0))
a_test.head()
```

```python
for i in range(0,len(a_test)):
    num = a_test['index'][i]
    hds_test.loc[num,'주구매시간'] = a['주구매시간'][i]
```
