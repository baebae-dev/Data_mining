#### 주구매시간 결측치 저녁구매비율별 평균으로 채우기

```python
new = hds_train[['주구매시간','저녁구매비율']]

nonan = new[pd.notnull(new['주구매시간']) == True]
nan = new[pd.notnull(new['주구매시간']) != True]
print(nonan.head())
print(nan.head())
```

```python
nonan_new = nonan.apply(lambda x: round(x,2)) ; nonan_new.head()
```

```python
stand = nonan_new.groupby(['저녁구매비율'])['주구매시간'].mean().reset_index() ; stand.head()
```

```python
stand_apply = hds_train[pd.notnull(hds_train['주구매시간']) != True]['저녁구매비율'].reset_index().apply(lambda x: round(x,2))
stand_apply.head()
```

```python
a = pd.merge(stand_apply,stand,on='저녁구매비율') ;a.head()
```

```python
import math
a['주구매시간'] = a['주구매시간'].apply(lambda x: math.ceil(x))
a.head(10)
```

```python
for i in range(0,len(a)):
    num = hds_train['custid'][a['index'][i]]
    hds_train.loc[num,'주구매시간'] = a['주구매시간'][i]
```
